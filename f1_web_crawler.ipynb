```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
url = 'https://www.formula1.com/en/results/2024/drivers'
response = requests.get(url)
```

```python
if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')
    table = soup.find('table',class_='f1-table f1-table-with-data w-full')
    #extract the rows of the table
    rows = table.find_all('tr')
    
    for row in rows:
        columns = row.find_all('td')
        if columns:
            pos = columns[0].text.strip()
            driver_raw = columns[1].text.strip()
            # Dùng regex để tách tên, họ và mã quốc gia
            match = re.match(r"(\w+)\s+([A-Za-z]+)([A-Z]{3})$", driver_raw)
            if match:
                first_name, last_name, code = match.groups()
                driver = [first_name, last_name]
                driver = ' '.join(driver) 
            else:
                driver = driver_raw.split()  # Dự phòng nếu không khớp mẫu

            nationality = columns[2].text.strip()
            car = columns[3].text.strip()
            print(f"Position: {pos} \
                    \nDriver Name: {driver}\
                    \nNationnality: {nationality}\
                    \nCar: {car} \n" )
         
```
